```python
import pandasql as psql
import plotly.express as px

import numpy as np
import math
from matplotlib import pyplot as plt

import pandas as pd
import seaborn as sns


import cv2
import os
from os.path import isfile, join,exists
```

```python
path = 'boneage-training-dataset.csv'
```

```python
DATA = pd.read_csv(path)
```

```python
consulta = """
        SELECT
            case male
                when True then 'Hombre'
                when False then 'Mujer'
            end as sexo
            ,count(male) as cantidad
        FROM
            DATA
        GROUP BY
            male
    """

datos = psql.sqldf(consulta, locals())
```

```python
fig = px.pie(datos, values='cantidad', names='sexo', title='Cantidad de Hombres y Mujeres en Set de Datos')
fig.show()
```

```python
consulta = """
        SELECT distinct
            boneage
        FROM
            DATA
        where
            male = True
    """

datos = psql.sqldf(consulta, locals())
```

```python
data = datos['boneage'].tolist()
```

```python
bins = np.linspace(math.ceil(min(data)), 
                   math.floor(max(data)),
                   20) # fixed number of bins

plt.xlim([min(data)-5, max(data)+5])

plt.hist(data, bins=bins, alpha=0.5)
plt.title('Histograma de Edad de Hueso en Hombres')
plt.xlabel('Edad')
plt.ylabel('Cantidad')

plt.show()
```

```python
data_hist = pd.read_csv("boneage-training-dataset.csv")
data_hist['boneage'].describe()
```

```python
sns.set_style("white")
sns.distplot(data_hist['boneage'],bins=30,kde=False, color="green", label="Compact")
plt.show()
```

```python
boneage = data_hist['boneage']
male = data_hist['male']

mujeres = []
hombres = []
```

```python
for x in range(0,len(boneage)):
    if male[x] == True:
        hombres.append(boneage[x])
    else:
        mujeres.append(boneage[x])
```

```python
sns.distplot(mujeres,bins=30,kde=False, color="red", label="Mujeres")
sns.distplot(hombres,bins=30,kde=False, color="blue", label="Hombres")
plt.legend();
plt.show()
```

```python
labels = 'Mujeres', 'Hombres'
sizes = [len(mujeres), len(hombres)]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
consulta = """
        SELECT distinct
            boneage
        FROM
            DATA
        where
            male = False
    """

datos = psql.sqldf(consulta, locals())
```

## Visualizando Proceso de Transformación de Imagen

```python
img = cv2.imread('{0}/{1}'.format('boneage-test-dataset','4360.png'))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
img_t = cv2.addWeighted(img,4, cv2.GaussianBlur(img , (0,0) , 30) ,-4 ,128)
edges = cv2.Canny(img_t,100,200)
        
f, axarr = plt.subplots(1,3,figsize = (22,22))
        
axarr[0].imshow(img)
axarr[0].set_title('Imagen Original')

axarr[1].imshow(img_t)
axarr[1].set_title('Imagen luego de Aclarar')

axarr[2].imshow(edges)
axarr[2].set_title('Bordes de la Imagen')
plt.show()
```
